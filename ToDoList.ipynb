```python
def add_task(tasks):
    while True:
        try:
            T1 = int(input('No. of Tasks to add (1-5): '))
            if 1 <= T1 <= 5:
                break
            else:
                print("Please enter a number between 1 and 5.")
        except ValueError:
            print("Invalid input. Please enter a number between 1 and 5.")
    
    for i in range(T1):
        task = input(f'Enter task {i + 1}: ')
        tasks.append({"task": task, "completed": False})
    
    print("Tasks added:")
    for task in tasks:
        print(task['task'])

# Implement other functions (delete_task, view_completed_tasks, etc.) similarly
def delete_task(tasks, deleted_tasks):
    if not tasks:
        print("No tasks available to delete.")
        return
    
    print("Current tasks:")
    for idx, task in enumerate(tasks):
        print(f"{idx + 1}. {task['task']}")
    
    try:
        task_to_delete = int(input("Select the task number to delete: ")) - 1
        if 0 <= task_to_delete < len(tasks):
            deleted_tasks.append(tasks.pop(task_to_delete))
            print("Task deleted successfully.")
        else:
            print("Invalid task number.")
    except ValueError:
        print("Please enter a valid number.")

def view_completed_tasks(completed_tasks):
    if not completed_tasks:
        print("No completed tasks.")
    else:
        print("Completed tasks:")
        for task in completed_tasks:
            print(task['task'])

def view_remaining_tasks(tasks):
    if not tasks:
        print("No remaining tasks.")
    else:
        print("Remaining tasks:")
        for task in tasks:
            print(task['task'])

def view_deleted_tasks(deleted_tasks):
    if not deleted_tasks:
        print("No deleted tasks.")
    else:
        print("Deleted tasks:")
        for task in deleted_tasks:
            print(task['task'])

def main():
    tasks = []
    completed_tasks = []
    deleted_tasks = []

    while True:
        print("\nQ. Select appropriate option")
        options = (
            "1. Add task",
            "2. Delete task",
            "3. View all completed tasks",
            "4. View all remaining tasks",
            "5. View deleted list of tasks",
            "6. Exit the program"
        )
        
        for option in options:
            print(option)

        try:
            answer = int(input("Choose an option: "))
        except ValueError:
            print("Please enter a number between 1 and 6.")
            continue
        
        if answer == 1:
            add_task(tasks)
        elif answer == 2:
            delete_task(tasks, deleted_tasks)
        elif answer == 3:
            view_completed_tasks(completed_tasks)
        elif answer == 4:
            view_remaining_tasks(tasks)
        elif answer == 5:
            view_deleted_tasks(deleted_tasks)
        elif answer == 6:
            print("Exiting the program")
            break
        else:
            print("Your answer is wrong. Please choose a valid option.")

if __name__ == "__main__":
    main()
```
